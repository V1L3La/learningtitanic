# prever quais passageiros sobreviveriam ao terrível desastre
#####PassengerId: Número de identificação do passageiro;
#####Survived: Indica se o passageiro sobreviveu ao desastre. É atribuído o valor de 0 para aqueles que não sobreviveram, e 1 para quem sobreviveu;
#####Pclass: Classe na qual o passageiro viajou. É informado 1 para primeira classe; 2 para segunda; e 3 para terceira;
#####Name: Nome do passageiro;
#####Sex: Sexo do passageiro;
#####Age: Idade do passageiro em anos;
#####SibSp: Quantidade de (irmãos e cônjuges) a bordo ;
#####Parch: Quantidade de (pais e filhos) a bordo;
#####Ticket: Número da passagem;
#####Fare: (Tarifa)Preço da passagem;
#####Cabin: Número da cabine do passageiro;
#####Embarked: Indica o porto no qual o passageiro embarcou. Há apenas três valores possíveis: Cherbourg, Queenstown e Southampton, indicados pelas letras “C”, “Q” e “S”, respectivamente.

```python
import pandas as pd      #criar lindos DataFrames para visualizarmos nossos dados e trabalhar nosso código
from sklearn.tree import DecisionTreeClassifier   # responsável pela geração do modelo em si
```

```python
train = pd.read_csv('https://raw.githubusercontent.com/V1L3La/learningtitanic/main/train.csv')
train.head()
```

```python
train.head()
```

```python
test = pd.read_csv('https://raw.githubusercontent.com/V1L3La/learningtitanic/main/test.csv')
test.head()
```

```python
test.head()
```

```python
train.drop(['Name', 'Ticket', 'Cabin'], axis=1, inplace=True)   #axis=1 coluna inteira    inplace  - salva a alteração direto no dataset, inves de numa variável
test.drop(['Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
```

```python
train.head()
```

```python
test.head()
```

```python
# transformar todos dados em números
new_data_train=pd.get_dummies(train)    # get_dummies Transforma categorias em colunas binarias 
new_data_train.head()
```

```python
# transformar todos dados em números
new_data_test=pd.get_dummies(test)
new_data_test
```

```python
# Tratando valores nulos
# .isnull() Eetorna todos os valores nulos encontrados
# .sum() Soma todas as ocorrências e agrupa
# .sort_values ordena os dados. 
# ascending=False Ordenar do maior para o menor
# o dataset possui 177 valores inválidos na coluna Age
new_data_train.isnull().sum().sort_values(ascending=False).head(10)
```

```python
 new_data_train["Age"].fillna(new_data_train["Age"].mean(), inplace=True)
 new_data_train
```

```python
 new_data_test["Age"].fillna(new_data_test["Age"].mean(), inplace=True)
 new_data_test
```

```python
# Tratando valores nulos
# .isnull() Eetorna todos os valores nulos encontrados
# .sum() Soma todas as ocorrências e agrupa
# .sort_values ordena os dados. 
# ascending=False Ordenar do maior para o menor
# o dataset possui 177 valores inválidos na coluna Age
new_data_test.isnull().sum().sort_values(ascending=False).head(10)
```

```python
 new_data_test["Fare"].fillna(new_data_test["Fare"].mean(), inplace=True)
```

```python
# STOP in   Criando um modelo de Machine Learning
```

```python
X = new_data_train.drop('Survived', axis=1)
y = new_data_train['Survived']
```

```python
# cria modelo
tree = DecisionTreeClassifier(max_depth=3, random_state=0)
tree.fit(X,y) 
                       
                              
```

```python
tree.score(X,y)
```

```python
# Enviando primeira previsão ao Kaggle em .csv (Parte I)
submission = pd.DataFrame()
submission['PassengerId'] = new_data_test['PassengerId']
submission['Survived'] = tree.predict(new_data_test)
```

```python
submission
```

```python
# Enviando primeira previsão ao Kaggle em .csv (Parte II)
submission.to_csv('submission_final.csv', index=False)
```
